# Data Penjualan Minimarket


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url_maret = 'https://raw.githubusercontent.com/angelaglo/minimarket_sales_report/main/laporan_maret.csv'
url_april = 'https://raw.githubusercontent.com/angelaglo/minimarket_sales_report/main/laporan_april.csv'

laporan_maret = pd.read_csv(url_maret, encoding= 'unicode_escape')
laporan_april = pd.read_csv(url_april, encoding= 'unicode_escape')
```

```python
laporan_maret
```

```python
laporan_april
```

# Pengecekan data dan cleaning data

```python
# Maret
laporan_maret.isnull().sum()
```

```python
# April
laporan_april.isnull().sum()
```

```python
# membersihkan dari nilai NaN
laporan_april.dropna(axis=1, how="all", inplace=True)
laporan_april
```

```python
# menghilangkan spasi pada nama kolom agar dapat di akses
laporan_maret.columns = laporan_maret.columns.str.strip()
laporan_april.columns = laporan_april.columns.str.strip()
```

```python
# pengecekan tipe data maret
laporan_maret.dtypes
```

```python
# pengecekan tipe data april
laporan_april.dtypes
```

```python
#Format "Harga (Per PCS)" dan "Sub Total" pada bulan Maret disamakan dengan bulan April dengan menghilangkan mata uang dan koma
laporan_maret["Harga (Per PCS)"] = laporan_maret['Harga (Per PCS)'].str.replace(',','').str.replace('Rp','').astype(int)
laporan_maret["Sub Total"] = laporan_maret['Sub Total'].str.replace(',', '').str.replace('Rp', '').astype(int)
laporan_maret
```

```python
# merubah tipe data Tanggal
laporan_maret['Tanggal Transaksi'] = pd.to_datetime(laporan_maret['Tanggal Transaksi'])
laporan_maret
```

```python
laporan_april['Tanggal Transaksi'] = pd.to_datetime(laporan_april['Tanggal Transaksi'])
laporan_april['Tanggal Transaksi'] = laporan_april['Tanggal Transaksi'].dt.date
laporan_april['Tanggal Transaksi'] = pd.to_datetime(laporan_april['Tanggal Transaksi'])
laporan_april
```

```python
laporan_april.dtypes
```

```python
laporan_maret.dtypes
```

```python
# penggabungan laporan
frames = [laporan_maret, laporan_april]

laporan_all = pd.concat(frames, ignore_index=True)
laporan_all
```

# DATA PROSESING

```python
# Fungsi untuk grafik pendapatan

def grafikPendapatan(x, y) :
  f = plt.figure()
  f.set_figwidth(15)
  f.set_figheight(5)
  plt.title("Grafik Pendapatan")
  plt.plot(x, y, 'o-')
  plt.ticklabel_format(axis='y', style='plain')
  plt.xlabel('Tanggal')
  plt.ylabel('Pendapatan')
  plt.xticks(x,  rotation=90)


  plt.show()
```

```python
#Laporan Keseluruhan Bulan Maret dan April
all_report = laporan_all.copy()
all_produk = all_report.groupby("Produk")
rangkuman_all = all_produk["Produk"].count()\
                                    .reset_index(name="count")\
                                    .sort_values(["count"],\
                                                ascending=False)

#Menentukan produk terlaris dan tidak
max_all = rangkuman_all.head(15)
min_all = rangkuman_all.tail(15)

print("============== PRODUK TERLARIS 15 TERATAS ================")
print(max_all.to_string(index=False), "\n")
print("\n============== PRODUK KURANG LAKU 15 TERBAWAH ================")
print(min_all.to_string(index=False), "\n")

#Menampilkan jumlah transaksi per hari
tanggal_all = all_report.groupby("Tanggal Transaksi")
rangkuman_tanggal_all = tanggal_all["Tanggal Transaksi"].count()

print(" GRAFIK TOTAL PENJUALAN ".center(100, '='))
tanggal_all["Tanggal Transaksi"].hist(xrot = 90)
print(rangkuman_tanggal_all)

#Menampilkan jumlah pendapatan per hari
pendapatan_all = all_report["Sub Total"].sum()
groupsum_all = all_report.groupby(by="Tanggal Transaksi", as_index=False)["Sub Total"].sum()

grafikPendapatan(groupsum_all['Tanggal Transaksi'], groupsum_all['Sub Total'])
print('Total Pendapatan Keseluruhan : Rp ', pendapatan_all)
```

```python
#data processing bulan april
all_april = laporan_april.copy()
produk_april = all_april.groupby("Produk")
rangkuman_april = produk_april["Produk"].count()\
                                        .reset_index(name="count")\
                                        .sort_values(["count"],\
                                                     ascending=False)
max_april = rangkuman_april.head(15)
min_april = rangkuman_april.tail(15)

tanggal_april = all_april.groupby("Tanggal Transaksi")
rangkuman_tanggal_april = tanggal_april["Tanggal Transaksi"].count()

pendapatan_april = all_april["Sub Total"].sum()
groupsum_april = all_april.groupby(by="Tanggal Transaksi", as_index=False)["Sub Total"].sum()
```
